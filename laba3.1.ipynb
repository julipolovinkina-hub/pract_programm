```python
# Задание task_03_01_03.
#
# Выполнил: Половинкина Ю. С.
# Группа: ЦИБ-251


def is_lucky(num):
    """Является ли 'num' номером счастливого билета?

    Параметры:
        num (int): Номер билета для проверки.

    Результат:
        bool: True - да, False - нет.
    """
    even_count = 0
    odd_count = 0

    n = num
    while n > 0:
        d = n % 10
        if d % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
        n = n // 10

    return even_count == odd_count


def lucky_numbers(a, b):
    """Вернуть список счастливых номеров от 'a' до 'b'.

    Параметры:
        a (int): Начало диапазона (включительно).
        b (int): Конец диапазона (включительно).

    Результат:
        list: Список счастливых номеров в заданном диапазоне.
    """
    lucky_list = []
    for i in range(a, b + 1):
        if is_lucky(i):
            lucky_list.append(i)

    return lucky_list


a = int(input("Первый номер билета: "))
b = int(input("Последний номер билета: "))

# Вывод должен быть осуществлен в строку с одним пробелом-разделителем
result = lucky_numbers(a, b)

for i in range(len(result)):
    if i == 0:
        print(result[i], end='')
    else:
        print(' ' + str(result[i]), end='')
print()

# --------------
# Пример вывода:
#
# Первый номер билета: 10
# Последний номер билета: 25
# 10 12 14 16 18 21 23 25
```

Для поиска «счастливых» билетов в диапазоне от *a* до *b* программа использует две  функции: `is_lucky` арифметическим путём (через операции `% 10` и `// 10`) извлекает цифры числа, подсчитывает чётные и нечётные и возвращает результат их сравнения, а `lucky_numbers` перебирает заданный диапазон и собирает подходящие номера в список, который затем считывается и выводится основным блоком кода в одну строку через пробел.
