```python
# Задание task_03_03_10.
#
# Выполнил: Половинкина Ю. С.
# Группа: ЦИБ-251

"""
Ошибки (номера строк через пробел, данная строка - №2): 34 35 46
"""


# В программе хранится словарь вступительных экзаменов вида Предмет=Баллы.
# Узнать, проходит ли абитуриент на специальность, если нужно сдать
# все перечисленные экзамены с не меньшим количеством баллов.

необходимые_экзамены = {
    "Информатика": 80,
    "Математика": 85,
    "Русский язык": 75
}

print("""Для определения возможности поступления, необходима информация о Вас.

Для ввода экзамена и баллов введите их через |: Химия | 40.
Для завершения ввода нажмите Enter.
""")

сданные_экзамены = {}
while True:
    try:
        ввод = input("").strip()
        if ввод == "":
            break

        экзамен, балл = [x.strip() for x in ввод.split("|")]
        сданные_экзамены[экзамен] = int(балл)
    except ValueError:
        print("Ошибка: введите данные в формате 'Экзамен | Баллы' (баллы должны быть числом).")

print("Ваши экзамены:")
for i, (экзамен, балл) in enumerate(сданные_экзамены.items(), start=1):
    print("{}) {} {}".format(i, экзамен, балл))

ok = False
try:
    for необходимый_экзамен, баллы in необходимые_экзамены.items():
        if сданные_экзамены[необходимый_экзамен] < баллы:
            break
    else:
        ok = True
except KeyError:
    print("Ошибка: вы не сдали один из необходимых экзаменов.")

print("Вы можете к нам поступить!" if ok else "Увы...")
```

В строке 34 (экзамен, балл = [x.strip() for x in ввод.split("|")]) может возникнуть ValueError, если пользователь не ввёл разделитель | или ввёл больше/меньше двух значений.
В строке 35 (сданные_экзамены[экзамен] = int(балл)) - ValueError, если введённые баллы не являются целым числом.
В строке 46 (if сданные_экзамены[необходимый_экзамен] < баллы:) - KeyError, если абитуриент забыл ввести один из обязательных экзаменов. Благодаря конструкции try-except программа не завершается аварийно при некорректных данных, а выводит понятные сообщения об ошибках и продолжает работу.
