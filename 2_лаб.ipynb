```python
# Задание task_02_20.
#
# Выполнила: Половинкина Ю. С.
# Группа: ЦИБ-251

n = int(input())

for num in range(100, 1000):
    d1 = num // 100          # сотни
    d2 = (num // 10) % 10    # десятки
    d3 = num % 10            # единицы
    if d1 + d2 + d3 == n:
        print(num, end=' ')
print()                        # перевод строки после завершения
```
